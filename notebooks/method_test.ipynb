```python
import json
import numpy as np
from shapely.geometry import shape, box, GeometryCollection, Polygon
from shapely.affinity import scale
from packer import IrregularPackerIntensification, IrregularPackerStrictIntensification, IrregularPackerLSIntensification, IrregularPackerStrictLSIntensification, CirclePackerBeeIntensification, CirclePackerBeeStrictIntensification, CirclePackerBeeLSIntensification, CirclePackerBeeStrictLSIntensification
```

```python
container = box(-2, -2, 4, 2)
```

```python
with open('../shapes/random_polygons.json', 'r') as f:
    shapes = json.load(f)
    shapes = [shape(s) for s in shapes]
```

```python
radius = 0.5
```

```python
packer = IrregularPackerIntensification(container, shapes, n_rotations=1, n_search=10)
df = packer.pack()
print(packer.score())
GeometryCollection([container, *df['item']])
```

```python
packer = IrregularPackerStrictIntensification(container, shapes, n_rotations=1, n_search=10)
df = packer.pack()
print(packer.score())
GeometryCollection([container, *df['item']])
```

```python
packer = IrregularPackerLSIntensification(container, shapes, n_rotations=1, n_search=10)
df = packer.pack()
print(packer.score())
GeometryCollection([container, *df['item']])
```

```python
packer = IrregularPackerStrictLSIntensification(container, shapes, n_rotations=1, n_search=10)
df = packer.pack()
print(packer.score())
GeometryCollection([container, *df['item']])
```

```python
packer = CirclePackerBeeIntensification(container, radius, n_search=10)
df = packer.pack()
print('Number of circles:', len(df), '- Score:', round(packer.score(), 3))
GeometryCollection([container, *df['item']])
```

```python
packer = CirclePackerBeeStrictIntensification(container, radius, n_search=10)
df = packer.pack()
print('Number of circles:', len(df), '- Score:', round(packer.score(), 3))
GeometryCollection([container, *df['item']])
```

```python
packer = CirclePackerBeeLSIntensification(container, radius, n_search=10)
df = packer.pack()
print('Number of circles:', len(df), '- Score:', round(packer.score(), 3))
GeometryCollection([container, *df['item']])
```
