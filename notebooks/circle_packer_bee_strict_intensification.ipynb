```python
from shapely.geometry import GeometryCollection

from packer import CirclePackerBeeStrictIntensification
from tests import ALL_TESTS
```

```python
container = ALL_TESTS[0]['container']
radius = ALL_TESTS[0]['radius']
```

```python
%%time
packer = CirclePackerBeeStrictIntensification(
    container,
    radius,
    intersection_threshold=40,
    max_iter=1000,
    shots=5,
    n_neighbors=8,
    step_portion=2,
    queue_length=3,
    n_search=10
)
df_sol = packer.pack()
```

```python
print('Number of circles:', len(df_sol), '- Score:', round(packer.score(), 3))
GeometryCollection([container, *df_sol['item']])
```
